```python
import pandas as pd
```

```python
#Load lightweight data
file_path = "LW_clean_data.csv"
lw_data_df = pd.read_csv(file_path)
lw_data_df
```

```python
#Create Red fighter df
r_df = lw_data_df.drop(lw_data_df.loc[:, "R_fighter":"B_Weight_lbs"].columns, axis = 1)
r_df = r_df.drop(r_df.columns[[0,]],axis=1)
r_df = r_df.drop(["B_age"],axis =1)
r_df = r_df.drop(r_df.loc[:, "B_Stance_Orthodox":"Winner_Draw"].columns, axis = 1)
r_df.head()
```

```python
#Create Blue fighter df
b_df = lw_data_df.drop(lw_data_df.loc[:, "R_avg_KD":"R_Weight_lbs"].columns, axis = 1)
b_df = b_df.drop(b_df.loc[:, "R_fighter":"B_fighter"].columns, axis = 1)
b_df = b_df.drop(b_df.columns[[0,]],axis=1)
b_df = b_df.drop(lw_data_df.loc[:, "R_age":"R_Stance_nan"].columns, axis = 1)
b_df = b_df.drop(["Winner_Red", "Winner_Draw"], axis=1)
b_df
```
