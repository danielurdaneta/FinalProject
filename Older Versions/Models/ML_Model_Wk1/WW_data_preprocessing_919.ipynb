```python
# Import our dependencies
import pandas as pd
from sklearn.preprocessing import StandardScaler,OneHotEncoder, minmax_scale

```

```python
data_df = pd.read_csv("WW_clean_data_labelEncoded_Winner.csv")
data_df
```

```python
# Create the OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit the encoder and produce encoded DataFrame
B_Country = pd.DataFrame(enc.fit_transform(data_df.B_Country.values.reshape(-1,1)))

# Rename encoded columns
B_Country.columns = enc.get_feature_names(['B_Country'])
B_Country
```

```python
# Create the OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit the encoder and produce encoded DataFrame
R_Country = pd.DataFrame(enc.fit_transform(data_df.R_Country.values.reshape(-1,1)))

# Rename encoded columns
R_Country.columns = enc.get_feature_names(['R_Country'])
R_Country
```

```python
#Merge one hot encoded country data
merged_df=data_df.join(B_Country)
merged_df2=merged_df.join(R_Country)

merged_df2.head()
```

```python
merged_df2=merged_df2.drop(['Unnamed: 0'], axis=1)
merged_df2.head()
```

```python
from sklearn.preprocessing import minmax_scale

merged_df2[['B_total_time_fought(seconds)','R_total_time_fought(seconds)', 'B_Height_cms', 'R_Height_cms', 'B_Reach_cms', 'R_Reach_cms', 'B_Weight_lbs', 'R_Weight_lbs']] = minmax_scale(merged_df2[['B_total_time_fought(seconds)','R_total_time_fought(seconds)', 'B_Height_cms', 'R_Height_cms', 'B_Reach_cms', 'R_Reach_cms', 'B_Weight_lbs', 'R_Weight_lbs']])
```

```python
merged_df2=merged_df2.drop(['B_Country', 'R_Country'], axis=1)
merged_df2.head()
```

```python
merged_df2.to_csv('WW_clean_data_labelEncoded_Winner.csv', index=False)
```
