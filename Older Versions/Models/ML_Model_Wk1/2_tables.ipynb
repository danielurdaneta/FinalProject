```python
# Import our dependencies
import pandas as pd
import numpy as np
```

```python
data_df = pd.read_csv("WW_clean_data_labelEncoded_Winner.csv")
data_df
```

```python
#Split into the first dataframe for just winner & fighter names & add fight ID
df1 = data_df.iloc[:, :4]
df1
```

```python
#Call list to insert fight Ids
fight_ids = np.arange(1,114,1).tolist()
fight_ids
```

```python
#Insert fight ID column
df1.insert(loc=0,column = "fightID", value = fight_ids)
df1
```

```python
#Split into the first dataframe for just winner & fighter names
df2 = data_df.iloc[:, 4:]
df2
```

```python
#Insert fight ID column
df2.insert(loc=0,column = "fightID", value = fight_ids)
df2
```

```python
#Convert to csv files
df1.to_csv("fighters.winners.csv",index=False)
df2.to_csv("fighters.historicals.csv", index=False)
```
